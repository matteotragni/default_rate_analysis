 # Black-Scholes Model


```python
import numpy as np
import scipy.stats as stat
```

```python
# Variables
r = 0.01
S = 30
K = 40
T = 240/365
sigma = 0.3
```

```python
def BlackScholes(r, S, K, T, sigma, type="C"):
    "BlackScholes option price for a call/put"
    d1 = (np.log(S/K)+ (r + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    try: 
        if type =="C":
            price = S*stat.norm.cdf(d1, 0, 1) - K*np.exp(-r*T)*stat.norm.cdf(d2, 0, 1)
    
        elif type == "P":
            price = K*np.exp(-r*T)*stat.norm.cdf(-d2, 0, 1) - S*stat.norm.cdf(-d1, 0, 1)
        return price
    
        
    except: 
        print("Please confirm all option parameters above!")

          
print("Call price: ", round(BlackScholes(r, S, K, T, sigma, type="C"),2))
print("Put price: ", round(BlackScholes(r, S, K, T, sigma, type="P"),2))
```
